```python
import requests
from bs4 import BeautifulSoup 

url = 'https://www.moviechart.co.kr/rank/realtime/index/image'
response = requests.get(url)
soup =  BeautifulSoup(response.text, 'html.parser')
soup
```

```python
#content > div.wArea.space > div.movieBox > ul > li:nth-child(4) > div > div.movie-title > h3 > a
movie_list = soup.select('div.movieBox div.movie-title  a ')
for i in movie_list:
    print(i.text)
```

```python
from PIL import Image
from io import BytesIO
```

```python
img_dir = "images"

import os
if not os.path.exists(img_dir):
    os.makedirs(img_dir)

full_image_dir = os.path.join(os.getcwd(), img_dir)
full_image_dir
```

```python
#content > div.wArea.space > div.movieBox > ul > li:nth-child(4) > a > img
#https://www.moviechart.co.kr/thumb?width=178&height=267&m_code=20242248&source=https://www.moviechart.co.kr/assets/upload/movie/241023101033_3849.jpg
image_list = soup.select('div.movieBox  a  img')
for i in image_list:
   img_response = requests.get('https://www.moviechart.co.kr/' +  i['src'])
   #print(img_response.content)
   image = Image.open(BytesIO(img_response.content))
   image.save(os.path.join(img_dir, i['alt']+'.jpg')
```
